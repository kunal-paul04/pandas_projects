## Bar Plot

```python
# By using sns.barplot(), we can get different statistical aggregations like MEAN, MEDIAN, SUM, COUNT
```

1. Create a basic BARPLOT

```python
plt.figure(figsize=(12,4))
sns.barplot(x='Product line', y='Total', data=df)
```

2. Add hue to barplot

```python
plt.figure(figsize=(12,4))
sns.barplot(x='Product line', y='Total', hue='Gender', data=df)
```

3. Make the barplot Horizontal

```python
plt.figure(figsize=(10,4))
sns.barplot(x='Total', y='Product line', hue='Gender', data=df)
```

4. Plot the bars in given order - alphabetical order of Product line

```python
# First sort the category in albhabetical order
sorted_category = df['Product line'].sort_values().unique()
hue_sorted_order = df['Gender'].sort_values().unique()
plt.figure(figsize=(12,4))
sns.barplot(x='Product line', y='Total', hue='Gender', data=df, order=sorted_category, hue_order=hue_sorted_order)
```

5. Add CAP in the ERROR BAR
(Error Bar is the black straight line in each plot box)

```python
plt.figure(figsize=(12,4))
sns.barplot(x='Product line', y='Total', hue='Gender', data=df, order=sorted_category, hue_order=hue_sorted_order, capsize=0.1)
```

6. Remove the ERROR BAR using ci/errorbar (add new parameter - ci/errorbar=None {The `ci` parameter is deprecated. Use `errorbar=None` for the same effect.})

```python
plt.figure(figsize=(12,4))
sns.barplot(x='Product line', y='Total', hue='Gender', data=df, order=sorted_category, hue_order=hue_sorted_order, errorbar=None)
```

7. Change color using PALETTE

```python
plt.figure(figsize=(12,4))
sns.barplot(x='Product line', y='Total', hue='Gender', data=df, order=sorted_category, hue_order=hue_sorted_order, capsize=0.1, palette='cividis')
```

8. Use SATURATION prameter (add new parameter - saturation={value})

```python
plt.figure(figsize=(12,4))
sns.barplot(x='Product line', y='Total', hue='Gender', data=df, order=sorted_category, hue_order=hue_sorted_order, capsize=0.1, palette='cividis', saturation=70)
```

9. Change default aggregation method using ESTIMATOR parameter  (add new parameter - estimator={aggregation method - sum,median<np.median>,etc})
(By default, the barplot gives MEAN if you do not provide any other aggregation method)

```python
plt.figure(figsize=(12,4))
sns.barplot(x='Product line', y='Total', hue='Gender', data=df, order=sorted_category, hue_order=hue_sorted_order, capsize=0.1, estimator=np.median)
```
