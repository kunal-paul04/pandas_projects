## Count Plot

1. Create a countplot and analyze customer gender

```python
sns.countplot(x='Gender', data=df)
```

2. Create a countplot to get number of transaction for each of the product line

```python
# countplot with vertical bar plot
plt.figure(figsize=(12,4))
# used plt.figure() to enlarge the figure as the different categories of product line were not readable
sns.countplot(x='Product line', data=df)
```

```python
# countplot with horizontal bar plot
sns.countplot(y='Product line', data=df)
```

3. Add hue to get the count of to categories i.e., Product line and Gender

```python
plt.figure(figsize=(12,4))
sns.countplot(x='Product line', hue='Gender', data=df)
```
