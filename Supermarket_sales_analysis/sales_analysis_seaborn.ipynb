## Bar Plot

```python
# By using sns.barplot(), we can get different statistical aggregations like MEAN, MEDIAN, SUM, COUNT
```

1. Create a basic BARPLOT

```python
plt.figure(figsize=(12,4))
sns.barplot(x='Product line', y='Total', data=df)
```

2. Add hue to barplot

```python
plt.figure(figsize=(12,4))
sns.barplot(x='Product line', y='Total', hue='Gender', data=df)
```

3. Make the barplot Horizontal

```python
plt.figure(figsize=(10,4))
sns.barplot(x='Total', y='Product line', hue='Gender', data=df)
```

4. Plot the bars in given order - alphabetical order of Product line

```python
# First sort the category in albhabetical order
sorted_category = df['Product line'].sort_values().unique()
hue_sorted_order = df['Gender'].sort_values().unique()
plt.figure(figsize=(12,4))
sns.barplot(x='Product line', y='Total', hue='Gender', data=df, order=sorted_category, hue_order=hue_sorted_order)
```

5. Add CAP in the ERROR BAR
(Error Bar is the black straight line in each plot box)

```python
plt.figure(figsize=(12,4))
sns.barplot(x='Product line', y='Total', hue='Gender', data=df, order=sorted_category, hue_order=hue_sorted_order, capsize=0.1)
```

6. Remove the ERROR BAR using ci/errorbar (add new parameter - ci/errorbar=None {The `ci` parameter is deprecated. Use `errorbar=None` for the same effect.})

```python
plt.figure(figsize=(12,4))
sns.barplot(x='Product line', y='Total', hue='Gender', data=df, order=sorted_category, hue_order=hue_sorted_order, errorbar=None)
```

7. Change color using PALETTE

```python
plt.figure(figsize=(12,4))
sns.barplot(x='Product line', y='Total', hue='Gender', data=df, order=sorted_category, hue_order=hue_sorted_order, capsize=0.1, palette='cividis')
```

8. Use SATURATION prameter (add new parameter - saturation={value})

```python
plt.figure(figsize=(12,4))
sns.barplot(x='Product line', y='Total', hue='Gender', data=df, order=sorted_category, hue_order=hue_sorted_order, capsize=0.1, palette='cividis', saturation=70)
```

9. Change default aggregation method using ESTIMATOR parameter  (add new parameter - estimator={aggregation method - sum,median<np.median>,etc})
(By default, the barplot gives MEAN if you do not provide any other aggregation method)

```python
plt.figure(figsize=(12,4))
sns.barplot(x='Product line', y='Total', hue='Gender', data=df, order=sorted_category, hue_order=hue_sorted_order, capsize=0.1, estimator=np.median)
```

# Box Plot

```python
# This provides a summary of supplied data, which includes the information like: Mean, Median, Minimum, Maximum, 1st Quartile, 3rd Quartile, Outliers
```

1. Create a basic boxplot on a numerical variable by a categorical variable

```python
sns.boxplot(x='Payment', y='Total', data=df, width=0.5)
```

2. Create a basic boxplot on a numerical variable by two categorical variables using hue

```python
sns.boxplot(x='Payment', y='Total', data=df, hue='Gender', width=0.5)
```

3. Add mean marker in the boxplot using Showmeans and change its style using meanprops

```python
sns.boxplot(x='Payment', y='Total', data=df, hue='Gender', width=0.5, showmeans=True, meanprops={"marker":"o",
                                                                                                 "markerfacecolor":"white",
                                                                                                 "markersize":"10",
                                                                                                 "markeredgecolor":"black"}
           , palette="Set2", linewidth=2)
```

# Violin Plot

1. Create a basic violin plot

```python
sns.violinplot(x='Payment', y='Total', hue='Gender', data=df, split=True, inner="quartile", bw_method=0.2)
# bw is use for smoothning the lines and inner is use to mark Mean, Median, Minimum, Maximum, 1st Quartile, 3rd Quartile, Outliers
```

# Strip Plot

```python
# A strip plot is a graphical data analysis technique for summerizing a univariate data set
# A strip plot can be drawn on its own, but is is also a good complement to a box or violin plot in case where you want to show all observations along with some representation of the underlying distribution
```

1. Create a basic strip plot

```python
sns.stripplot(data=df, y='Total', linewidth=0.5)
```

```python
sns.stripplot(data=df, y='Total', jitter=0.4, linewidth=0.5)
```

2. Include third category using hue

```python
sns.stripplot(data=df, x='Payment', y='Total', hue='Gender', jitter=0.2, linewidth=0.5)
```

```python
sns.stripplot(data=df, x='Payment', y='Total', hue='Gender', jitter=0.2, linewidth=0.5, dodge=True)
```

3. Draw the strip plot on top of box plot

```python
sns.stripplot(data=df, x='Payment', y='Total', hue='Gender', jitter=0.2, linewidth=0.5)
sns.boxplot(data=df, x='Payment', y='Total', color='grey')
```
