# Importing necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

# Load CSV

```python
df = pd.read_csv('hotel_bookings.csv')
```

```python
# Convert any column having date to datetime format
# Convert 'reservation_status_date' to datetime format, allowing for mixed date formats and using 'dayfirst=True'
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'], dayfirst=True, errors='coerce')
df.info()
```

```python
# Show detailes analysis for columns that are object
for col in df.describe(include = 'object').columns:
    # Show columns that are object
    print(col)
    # Show unique values in these columns
    print(df[col].unique())
    # Add ---- after each column record to separate data
    print('-'*50)
```

```python
# Show columns that is missing record along with its count
df.isnull().sum()
```

```python
# Drop columns are not required and that have more missing values
# To remove column, put axis = 1
# To change the dataframe, put inplace = True
df.drop(['company', 'agent'], axis = 1, inplace = True)
# Remove rows having missing values
df.dropna(inplace = True)
# Now check missing columns
df.isnull().sum()
```

```python
# Check record description after changes
df.describe()
```

```python
# Since there is outlier in 'adr', remove outlier value from adr
df = df[df['adr'] < 5000]
# Check record description after changes
df.describe()
```

# Since data is clean, analysis data - Data Analysis and visualization

```python
# Check data count for reservation that are cancelled and not cancelled
# Analyze the reservation cancellation. The data shows for all hotels.
reservation_cancellation_rec = df['is_canceled'].value_counts(normalize = True)
print(reservation_cancellation_rec)
print('-'*50)
plt.figure(figsize = (5,4))
plt.title('Reservation Status')
plt.bar(['Not Cancelled', 'Cancelled'], df['is_canceled'].value_counts(), edgecolor = 'k', width = 0.7)
plt.show()
```

```python
# Analyze the above data according to respective hotels
plt.figure(figsize = (8,4))
ax1 = sns.countplot(x = 'hotel', hue = 'is_canceled', data = df, palette = 'Blues')
legend_labels, _ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor = (1,1))
plt.title('Reservation status in different hotels', size = 20)
plt.xlabel('Hotel')
plt.ylabel('Number of reservations')
plt.legend(['Not Cancelled', 'Cancelled'])
plt.show()
```
