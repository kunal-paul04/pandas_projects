# Importing necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

# Load CSV

```python
df = pd.read_csv('hotel_bookings.csv')
```

```python
# Convert any column having date to datetime format
# Convert 'reservation_status_date' to datetime format, allowing for mixed date formats and using 'dayfirst=True'
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'], dayfirst=True, errors='coerce')
df.info()
```

```python
# Show detailes analysis for columns that are object
for col in df.describe(include = 'object').columns:
    # Show columns that are object
    print(col)
    # Show unique values in these columns
    print(df[col].unique())
    # Add ---- after each column record to separate data
    print('-'*50)
```

```python
# Show columns that is missing record along with its count
df.isnull().sum()
```

```python
# Drop columns are not required and that have more missing values
# To remove column, put axis = 1
# To change the dataframe, put inplace = True
df.drop(['company', 'agent'], axis = 1, inplace = True)
# Remove rows having missing values
df.dropna(inplace = True)
# Now check missing columns
df.isnull().sum()
```

```python
# Check record description after changes
df.describe()
```

```python
# Since there is outlier in 'adr', remove outlier value from adr
df = df[df['adr'] < 5000]
# Check record description after changes
df.describe()
```

# Since data is clean, analysis data - Data Analysis and visualization

```python
# Check data count for reservation that are cancelled and not cancelled
# Analyze the reservation cancellation. The data shows for all hotels.
reservation_cancellation_rec = df['is_canceled'].value_counts(normalize = True)
print(reservation_cancellation_rec)
print('-'*50)
plt.figure(figsize = (5,4))
plt.title('Reservation Status')
plt.bar(['Not Cancelled', 'Cancelled'], df['is_canceled'].value_counts(), edgecolor = 'k', width = 0.7)
plt.show()
```

```python
# Analyze the above data according to respective hotels
plt.figure(figsize = (8,4))
ax1 = sns.countplot(x = 'hotel', hue = 'is_canceled', data = df, palette = 'Blues')
legend_labels, _ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor = (1,1))
plt.title('Reservation status in different hotels', size = 20)
plt.xlabel('Hotel')
plt.ylabel('Number of reservations')
plt.legend(['Not Cancelled', 'Cancelled'])
plt.show()
```

```python
# Check cancellation % for individual hotel - Resort Hotel
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
# Check cancellation % for individual hotel - City Hotel
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
# Get average daily rate of individual hotels
resort_hotel_rate = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel_rate = city_hotel.groupby('reservation_status_date')[['adr']].mean()
# Generate graph
plt.figure(figsize = (20,8))
plt.title('Average Daily Rate in Hotels', fontsize = 30)
plt.plot(resort_hotel_rate.index, resort_hotel_rate['adr'], label = 'Resort Hotel')
plt.plot(city_hotel_rate.index, city_hotel_rate['adr'], label = 'City Hotel')
plt.legend(fontsize = 20)
plt.show()
```

```python
# Get reservation status per month
df['month'] = df['reservation_status_date'].dt.month
# Generate graph
plt.figure(figsize = (16,8))
ax1 = sns.countplot(x = 'month', hue = 'is_canceled', data = df, palette = 'bright')
legend_labels, _ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor = (1,1))
plt.title('Reservation status per month', size = 20)
plt.xlabel('month')
plt.ylabel('Number of reservations')
plt.legend(['Not Cancelled', 'Cancelled'])
plt.show()
```

```python
# Get average daily rate of hotel each month
plt.figure(figsize = (15,8))
plt.title('ADR per month', fontsize = 30)
sns.barplot(x='month', y='adr', data=df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())
plt.show()
```

```python
# Get top 10 countries with reservation cancelled
cancelled_data = df[df['is_canceled'] == 1]
top_ten_countries = cancelled_data['country'].value_counts()[:10]
plt.figure(figsize = (8,8))
plt.title('Top 10 countries with reservation cancelled')
plt.pie(top_ten_countries, autopct = '%.2f', labels = top_ten_countries.index)
plt.show()
```

```python
# Check from where clients are coming
print('Client Booking :-')
print(df['market_segment'].value_counts(normalize = True))
print('-'*50)
print('Client Cancellation :-')
print(cancelled_data['market_segment'].value_counts(normalize = True))
```

```python
# Analyze price of hotels that were booked and cancelled
booked_data = df[df['is_canceled'] == 0]
hotel_booked_rate = booked_data.groupby('reservation_status_date')[['adr']].mean()
hotel_booked_rate.reset_index(inplace = True)
hotel_booked_rate.sort_values('reservation_status_date', inplace = True)

hotel_cancelled_rate = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
hotel_cancelled_rate.reset_index(inplace = True)
hotel_cancelled_rate.sort_values('reservation_status_date', inplace = True)
# Generate graph
plt.figure(figsize = (20,6))
plt.title('Average Daily Rate', fontsize = 30)
plt.plot(hotel_booked_rate['reservation_status_date'], hotel_booked_rate['adr'], label = 'not cancelled')
plt.plot(hotel_cancelled_rate['reservation_status_date'], hotel_cancelled_rate['adr'], label = 'cancelled')
plt.legend(fontsize = 15)
plt.show()
```
