```python
mask_condition = csv_data['city'] == 'Hyderabad'
# -- mask_condition contains a boolean series for the condition that city == hyderabad
csv_data[mask_condition]
```

```python
# --> masking with multiple condition
condition_1 = csv_data['city'] == 'Hyderabad'
condition_2 = csv_data['season'] > 2015
csv_data[condition_1 & condition_2]
# using & operator in place of 'and' because condition_1 and condition_2 are boolean
```

```python
# value_count() returns the frequency count of data in a particular series/column
csv_data['winner'].value_counts()
```

```python
# for plot(), import matplotlib.pyplot as plot
# plot() creates a bar/graph of a particular series/column
# # Plot the winner counts
csv_data['winner'].value_counts().plot(kind='bar')
# # kind = bar | barh | pie | hist
```

```python
# for plot(), import matplotlib.pyplot as plot
# plot() creates a bar/graph of a particular series/column
# # Plot the winner counts
csv_data['winner'].value_counts().plot(kind='barh')
```

```python
# for plot(), import matplotlib.pyplot as plot
# plot() creates a bar/graph of a particular series/column
# # Plot the winner counts
csv_data['winner'].value_counts().plot(kind='pie')
```

```python
# for plot(), import matplotlib.pyplot as plot
# plot() creates a bar/graph of a particular series/column
# # Plot the winner counts
csv_data['winner'].value_counts().plot(kind='hist')
```

```python
# Checks for missing (NaN) values in the DataFrame
csv_data.isnull()
```

```python
# Apply a function along an axis (rows or columns).
csv_data['dl_applied'] = csv_data['dl_applied'].apply(lambda x: x + 5)
csv_data['dl_applied']
```

```python
# Groups the data by one or more columns and performs aggregation.
grouped = csv_data.groupby('winner').count()
grouped
```

```python
# Create a pivot table for summarizing data.
pivot = csv_data.pivot_table(values='win_by_wickets', index='winner', columns='win_by_runs', aggfunc='mean')
pivot
```
